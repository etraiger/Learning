```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import wikipedia
```

```python
dnvgl_colours = [
    '#009fda',  #cyan
    '#0f204b',  #dark blue
    '#36842d',  #land green
    '#e98300',  #orange
    '#fecb00',  #yellow
    '#6e5091',  #violet
    '#c4262e',  #red                                    
]

dnvgl_cm = ListedColormap(dnvgl_colours)
```

```python
# get text from wiki page on dnv gl
page = wikipedia.page('DNV_GL')
text = page.content
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# make the word cloud
wordcloud = WordCloud(font_path="windows\fonts\Avenir Next LT Pro\AvenirNextLTPro-Regular.otf",
                      max_font_size=60,                  
                      max_words = 100,
                      background_color="white", colormap=dnvgl_cm).generate(text)
```

```python
# show word cloud
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# save word cloud image
wordcloud.to_file("sample_wordcloud.jpg")
```
